```python
channels = ['красный', 'зеленый', 'синий']
for i in range(3):
    img_nuls = np.zeros(shape=(img.shape[0], img.shape[1], 3))
    img_nuls[:, :, i] = img[:, :, i]

    print "Изображение после зануления всех каналов, кроме канала: {}".format(channels[i])
    io.imshow(img_nuls)
    io.show()
    print "Изображение канала: {}".format(channels[i])
    io.imshow(img[:,:,i])
    io.show()
    
```

# 4

```python
mean_channels_img = img[:,:,:-1].mean(axis=2)

#пронормируем значения на отрезок [0,1] 
color_diff = mean_channels_img.max() - mean_channels_img.min()
normal_mean_channels_img = (mean_channels_img - mean_channels_img.min()) / color_diff

gray_img = skimage.color.rgb2gray(img)
concat_img = np.concatenate((normal_mean_channels_img, gray_img), axis=1)

print("Изображение по усредненным каналам | rgb2gray")
io.imshow(concat_img)
plt.show()
```

На мой взгляд, визуально чуть более четким получилось иображение, полученное путем усреднения каналов. Согласно документации, функция rgb2gray считает взвешенную сумму каналов(http://scikit-image.org/docs/dev/api/skimage.color.html#skimage.color.rgb2gray), причем вес зеленого канала наибольший. Из пункта 3 и блока ниже видно, что дисперсия у этого канала наименьшая, следовательно и изображение, где вес зеленого канала больше, будет менее четким.

```python
print "Канал : стандартное отклонение"
for i in range(3):
    print "{}: {}".format(channels[i], img[:, :, i].std())
```
